```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
X = pd.read_csv('/home/root1/Downloads/zoo_data(For Decision Tree Program).csv')
print(X)
```

```python
fn_names=['hair','feathers','eggs','milk','airborne','aquatic','predator','toothed','backbone','venomous','fins','legs','tail','domestic','catsize','types']
```

```python
X.head()
```

```python
y=X.values[:,-1];
```

```python
x=X.values[:,0:16]
```

```python
y
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=100)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
clf=DecisionTreeClassifier(criterion = "entropy",
            random_state = 100)
```

```python
clf.fit(x_train, y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
y_pred
```

```python
 print("Confusion Matrix: ",
        confusion_matrix(y_test, y_pred))
```

```python
 print ("Accuracy : ",
    accuracy_score(y_test,y_pred)*100)
```

```python
print("Report : ",
    classification_report(y_test, y_pred))
```

```python
cn=['1','2','3','4','5','6','7']
```

```python
plt.figure(figsize=(18,18))
tree.plot_tree(clf,feature_names=fn_names,class_names=cn,filled=True,fontsize=10)
```
